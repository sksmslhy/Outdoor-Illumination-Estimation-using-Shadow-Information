```python
def npz_data(npz):
    query = np.load(npz)
    return query['intrinsic_mat'], query['extrinsic_mat'], query['object_poses']
```

```python
query = np.load('./0/metadata/0100.npz')
O = query['object_poses'][2][1][:,3][:3]
S = query['object_poses'][4][1][:,3][:3]
```

```python
def cal_LD(path):
    query = np.load(path)
    O = query['object_poses'][2][1][:,3][:3] # Origin Position
    S = query['object_poses'][4][1][:,3][:3] # Sun Position
    v = S - O
    LD = v / np.sqrt(np.sum(v**2)) # Normal Vector (Light Direction)
    return  LD
```

```python
import os

path_dir = './dataset/0/metadata'
file_list = os.listdir(path_dir)

LD0 = []
for i in range(len(file_list)):
    path = path_dir + '/' + file_list[i]
    LD.append(cal_LD(path))
    
#print(np.array(LD).shape)

np.save('./light_direction0', LD)
```
