```python
df_train = pd.concat([df0,df1, df2,df3,df4,df5,df6], ignore_index=True)
df_test = pd.concat([df7,df8], ignore_index=True)
df_val = pd.concat([df9], ignore_index=True)
```

```python
for i in range(len(df_result)):
    df_result[' elevation'][i]  = df_result[' elevation'][i] * 180 / math.pi
    df_result[' azimuth'][i] = 90 - (df_result[' azimuth'][i] * 180 / math.pi)
```

```python
for i in range(len(df_train)):
    df_train[' elevation'][i]  = df_train[' elevation'][i] * 180 / math.pi
    df_train[' azimuth'][i] = 90 - (df_train[' azimuth'][i] * 180 / math.pi)
    
for i in range(len(df_test)):
    df_test[' elevation'][i]  = df_test[' elevation'][i] * 180 / math.pi
    df_test[' azimuth'][i] = 90 - (df_test[' azimuth'][i] * 180 / math.pi)
    
for i in range(len(df_val)):
    df_val[' elevation'][i]  = df_val[' elevation'][i] * 180 / math.pi
    df_val[' azimuth'][i] = 90 - (df_val[' azimuth'][i] * 180 / math.pi)
```

```python
df_train.to_csv("./train_labels.csv")
df_test.to_csv("./test_labels.csv")
df_val.to_csv("./val_labels.csv")
```
